# Remove outliers

```python
# Step 1: Define a function to remove outliers using IQR method
def remove_outliers(df, column):
    Q1 = df[column].quantile(0.25)
    Q3 = df[column].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    return df[(df[column] >= lower_bound) & (df[column] <= upper_bound)]

# Step 2: Apply the function to Price, Inches, and Weight columns

df_cleaned = remove_outliers(df_cleaned, 'Inches')
df_cleaned = remove_outliers(df_cleaned, 'Weight')

# Step 3: Optional - visualize boxplots for all three columns after removing outliers
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(18, 5))

for i, col in enumerate([ 'Inches', 'Weight']):
    plt.subplot(1, 3, i + 1)
    sns.boxplot(x=df_cleaned[col])
    plt.title(f'Boxplot of {col} (Cleaned)')

plt.tight_layout()
plt.show()
```

# Hypothesis Testing

```python
from scipy.stats import ttest_ind

# Assuming 'df' is your dataset and is already loaded
# Create two groups based on weight: lightweight (weight < 2 kg) and heavy (weight >= 2 kg)
lightweight = df[df['Weight'] < 2]['Price']
heavier = df[df['Weight'] >= 2]['Price']

# Perform independent T-test (Welch's T-test)
t_stat, p_value = ttest_ind(lightweight, heavier, equal_var=False)  # Welch's t-test for unequal variances

# Print results
print("T-statistic:", t_stat)
print("P-value:", p_value)

# Interpret the result
alpha = 0.05
if p_value < alpha:
    print("Reject the null hypothesis: Weight significantly difference in price.")
else:
    print("Fail to reject the null hypothesis: No significant difference in price.")
```

```python
from scipy.stats import ttest_ind

# Assuming 'df' is your dataset and is already loaded
ram_8 = df[df['RAM'] == 8]['Price']
ram_16 = df[df['RAM'] == 16]['Price']

# Perform independent t-test (Welch's t-test)
t_stat, p_value = ttest_ind(ram_8, ram_16, equal_var=False)

# Print test results
print("T-statistic:", t_stat)
print("P-value:", p_value)

# Interpret the result
alpha = 0.05
if p_value < alpha:
   print("Reject Hâ‚€: Price differs significantly based on RAM size.")
else:
    print("Fail to Reject Hâ‚€: No significant price difference based on RAM.")
```

```python
from scipy.stats import chi2_contingency

# Ensure columns are categorical strings
df['TypeName'] = df['TypeName'].astype(str)
df['Operating_System'] = df['Operating_System'].astype(str)

# Create a contingency table
contingency_table = pd.crosstab(df['TypeName'], df['Operating_System'])

# Perform Chi-Square Test
chi2_stat, p_val, dof, expected = chi2_contingency(contingency_table)

# Print the results
print("Chi-Square Test Results:")
print(f"Chi2 Statistic : {chi2_stat}")
print(f"P-value        : {p_val}")
print(f"Degrees of Freedom: {dof}")

# Hypothesis test conclusion
if p_val < 0.05:
    print("Conclusion: Reject Null Hypothesis â†’ Variables are dependent.")
else:
    print("Conclusion: Fail to Reject Null Hypothesis â†’ Variables are independent.")
```
